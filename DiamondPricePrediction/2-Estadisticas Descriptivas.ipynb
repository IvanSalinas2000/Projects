```python
df_copy= df.copy()
```

```python
#quitamos con drop los datos con dimenciones=0
df_copy = df_copy.drop(df_copy[df_copy["x"]==0].index)
df_copy = df_copy.drop(df_copy[df_copy["y"]==0].index)
df_copy = df_copy.drop(df_copy[df_copy["z"]==0].index)
df_copy.shape
```

```python
#validamos las filas donde hay 0
df_copy[df_copy.apply(lambda row: (row == 0).any(), axis=1)]
```

```python
df_copy.apply(lambda col: (col == 0).sum())
#podemos notar que ya no hay columnas con valores 0
```

```python
#vamos a quitar los valores anormales que excedan nuestros maximos y minimos
# Asumiendo que df_copy ya ha sido definido previamente
df_copy = df_copy[(df_copy["depth"] < 75) & (df_copy["depth"] > 45)]
df_copy = df_copy[(df_copy["table"] < 80) & (df_copy["table"] > 40)]
df_copy = df_copy[(df_copy["x"] < 30)]
df_copy = df_copy[(df_copy["y"] < 30)]
df_copy = df_copy[(df_copy["z"] < 30) & (df_copy["z"] > 2)]
df_copy.shape
```

```python
#Hacer una lista de las varibles categoricas
s = (df_copy.dtypes =="object")
object_cols = list(s[s].index)
print("Variable categoricas:")
print(object_cols)
```

```python
#Usamos LabelEncoder para hacer las categorias de las columnas cut, color y
#clarity para su clasificacion
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for col in object_cols:
    df_copy[col] = label_encoder.fit_transform(df_copy[col])
df_copy.head()
```

```python
df_copy.info()
```

```python
df_copy.tail()
```

```python
df_copy.reset_index(drop=True, inplace=True)
#Se reinician los indices
```

```python
df_copy.shape
```

```python
#Guardamos nuestro dataframe como csv
df_copy.to_csv("/content/diamantesCLEAN.csv")
```

Diagramas de dispersion

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Selecciona solo las columnas que deseas incluir en la matriz de dispersin
# Aqu se asume que quieres incluir todas las columnas del DataFrame
columns_of_interest = ["cut", "clarity", "table", "x", "y", "z", "carat", "color", "depth", "price"]
df_selected = df_diam[columns_of_interest]

# Crear el pair plot usando Seaborn
sns.pairplot(df_selected)
plt.savefig('matrizDispersion.png')
plt.show()
```

Diagrama de violin

```python
# Creamos una figura con subplots para cada columna
fig, axes = plt.subplots(nrows=len(columns_of_interest), ncols=1, figsize=(10, 4 * len(columns_of_interest)))

# Si slo hay una columna, el comportamiento de axes cambia, se hace necesario aadir esta condicin
if len(columns_of_interest) == 1:
    axes = [axes]

# Iterar sobre cada columna y crear un diagrama de violn
for ax, column in zip(axes, columns_of_interest):
    sns.violinplot(x=df_diam[column], ax=ax)
    ax.set_title(f'Diagrama de Violn de {column}')
    ax.set_xlabel(column)

plt.tight_layout()
plt.savefig('diagramas_violin.png')
plt.show()
```

Correlacion

```python
correlation_matrix = df_diam.corr()
# Suponiendo que correlation_matrix es tu matriz de correlacin
correlation_table = correlation_matrix.to_string()
# Imprimir la tabla completa
print(correlation_table)
```

```python
plt.figure(figsize=(10, 8))  # Ajusta el tamao de la figura si es necesario
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.savefig('correlacion.png')
plt.show()
```
